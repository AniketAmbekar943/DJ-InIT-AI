#### 22. Normalize a 5x5 random matrix (★☆☆)

````python
np.random.seed(50)
arr_2 = np.random.randint(1,10,(5,5))
# There are total three Types of Normalization
#1. Min-Max Normalization  
norm_1 = (arr_2-arr_2.min())/(arr_2.max()-arr_2.min())    #range [0,1] or [-1,1], best for neural networks, images
print("Min-Max Normalization:- \n",  norm_1, "\n")

#2. Z-score
norm_2 = (arr_2-arr_2.mean())/arr_2.std()                 #Scales data to mean=0 and std=1, good for static, linear model
print("Z- Score Normalization:- \n",  norm_2, "\n")

#3. vector normalization
norm_3 = arr_2/np.linalg.norm(arr_2)                    #returns a unti length vector, Distance based model
print("Vector Normalization:- \n",  norm_3, "\n")
````

#### 23. Create a custom dtype that describes a color as four unsigned bytes (RGBA) (★☆☆)

````python
import numpy as np

rgba = np.dtype([
    ('r', np.uint8),   #uint8 is 8 bit unsigned integer i.e. only +ve int 0-255
    ('g', np.uint8), 
    ('b', np.uint8),   
    ('a', np.uint8)    
])

print(rgba)

colors = np.array([(255, 0, 0, 255), (0, 255, 0, 128)], dtype=rgba)
print(colors)
print(colors['r'])  

````
