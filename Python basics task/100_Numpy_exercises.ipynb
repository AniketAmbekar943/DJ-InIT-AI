#### 22. Normalize a 5x5 random matrix (★☆☆)

````python
np.random.seed(50)
arr_2 = np.random.randint(1,10,(5,5))
# There are total three Types of Normalization
#1. Min-Max Normalization  
norm_1 = (arr_2-arr_2.min())/(arr_2.max()-arr_2.min())    #range [0,1] or [-1,1], best for neural networks, images
print("Min-Max Normalization:- \n",  norm_1, "\n")

#2. Z-score
norm_2 = (arr_2-arr_2.mean())/arr_2.std()                 #Scales data to mean=0 and std=1, good for static, linear model
print("Z- Score Normalization:- \n",  norm_2, "\n")

#3. vector normalization
norm_3 = arr_2/np.linalg.norm(arr_2)                    #returns a unti length vector, Distance based model
print("Vector Normalization:- \n",  norm_3, "\n")
````

#### 97. Considering 2 vectors A & B, write the einsum equivalent of inner, outer, sum, and mul function (★★★)

````python
A = np.arange(3)
B = np.arange(3,6)
#The einsum function is used to perform product of arrays.
inner = np.einsum('i,i->i', A, B) #the inner product of A and B is the sum of the products of the corresponding elements of A and B.
outer = np.einsum('i,j->ij', A, B) #the outer product of A and B is the product of the corresponding elements of A and B.
summation = np.einsum('i->', A) #the summation of A is the sum of the elements of A.
multiply = np.einsum('i,i->i', A, B)    #the product of A and B is the product of the corresponding elements of A and B.

inner, outer, summation, multiply
````
